```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
delta=linspace(57.1,59.9,100)
w1=60
w2=57
ratio1 = []
ratio2=[]
for i in delta:
    ratio1.append(2*(w1-i)/(w2-i))
    ratio2.append((w2-i)/(2*(w1-i)))
```

```python
plt.plot(delta,ratio1)
plt.grid()
plt.ylim([-5,0])
plt.show()
```

```python
plt.plot(delta,ratio2)
# plt.ylim([-5,0])
plt.grid()
plt.show()
```

```python
def res_freq(chi_rat,w1=w1,w2=w2):
    return (chi_rat*w2-2*w1)/(chi_rat-2)
```

```python
res_freq(6)
```
